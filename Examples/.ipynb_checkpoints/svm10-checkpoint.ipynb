```python
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score
from PIL import Image
from sklearn.preprocessing import LabelEncoder

def load_data(data_dir):
    X = []
    y = []
    file_paths = []
    for class_name in ['sipping_wine', 'smelling_wine', 'observing_wine']:
        class_dir = os.path.join(data_dir, class_name)
        for file_name in os.listdir(class_dir):
            file_path = os.path.join(class_dir, file_name)
            image = Image.open(file_path).convert('L')  # Convert to grayscale
            image = image.resize((64, 64))  # Resize to a fixed size
            image = np.array(image).flatten()  # Flatten the image
            X.append(image)
            y.append(class_name)
            file_paths.append(file_path)
    return np.array(X), np.array(y), file_paths

# Load dataset
data_dir = 'data'
X, y, file_paths = load_data(data_dir)

# Encode labels to integers
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)

# Initialize accumulators for results
test_accuracies = []
train_accuracies = []
test_reports = []
train_reports = []
results = []

# Run 10 times with different random splits
for i in range(10):
    # Split the dataset into training and test sets
    X_train, X_test, y_train, y_test, file_paths_train, file_paths_test = train_test_split(
        X, y, file_paths, test_size=0.2, random_state=i)

    # Standardize the features
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)

    # Apply PCA to reduce dimensionality
    pca = PCA(n_components=0.95)  # Keep 95% of variance
    X_train_pca = pca.fit_transform(X_train)
    X_test_pca = pca.transform(X_test)

    # Train an SVM classifier
    svm = SVC(kernel='linear', random_state=42)
    svm.fit(X_train_pca, y_train)

    # Predict on the test set
    y_pred_test = svm.predict(X_test_pca)
    # Predict on the training set
    y_pred_train = svm.predict(X_train_pca)

    # Evaluate the performance
    test_accuracy = accuracy_score(y_test, y_pred_test)
    train_accuracy = accuracy_score(y_train, y_pred_train)
    test_report = classification_report(y_test, y_pred_test, output_dict=True)
    train_report = classification_report(y_train, y_pred_train, output_dict=True)

    # Accumulate results
    test_accuracies.append(test_accuracy)
    train_accuracies.append(train_accuracy)
    test_reports.append(test_report)
    train_reports.append(train_report)

    # Extract recall and f1-score for the test set (macro average)
    recall_test = test_report['macro avg']['recall']
    f1_test = test_report['macro avg']['f1-score']
    results.append({
        'Run': i + 1,
        'Test Accuracy': test_accuracy,
        'Train Accuracy': train_accuracy,
        'Test Recall': recall_test,
        'Test F1 Score': f1_test
    })

# Calculate average results
avg_test_accuracy = np.mean(test_accuracies)
avg_train_accuracy = np.mean(train_accuracies)

# Print average results
print(f"Average Test Accuracy: {avg_test_accuracy}")
print(f"Average Train Accuracy: {avg_train_accuracy}")

# Calculate and print average classification reports
def average_classification_report(reports):
    avg_report = {}
    for key in reports[0].keys():
        if isinstance(reports[0][key], dict):
            avg_report[key] = {}
            for sub_key in reports[0][key].keys():
                avg_report[key][sub_key] = np.mean([report[key][sub_key] for report in reports])
        else:
            avg_report[key] = np.mean([report[key] for report in reports])
    return avg_report

avg_test_report = average_classification_report(test_reports)
avg_train_report = average_classification_report(train_reports)

print("Average Classification Report on test set:\n", avg_test_report)
print("Average Classification Report on training set:\n", avg_train_report)

# Create a DataFrame for the results and display it
results_df = pd.DataFrame(results)
print(results_df)

# Optionally, plot some test images with their predicted labels for the last iteration
fig, axes = plt.subplots(2, 3, figsize=(10, 5))
for i, ax in enumerate(axes.flat):
    image = X_test[i].reshape(64, 64)  # Reshape image back to 64x64
    ax.imshow(image, cmap='gray')
    true_label = label_encoder.inverse_transform([y_test[i]])[0]
    pred_label = label_encoder.inverse_transform([y_pred_test[i]])[0]
    ax.set_title(f"True: {true_label}\nPred: {pred_label}")
    ax.axis('off')
plt.show()

```
